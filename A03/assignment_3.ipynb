```python
# Access your secret keys via
from google.colab import userdata
# The name of your secret must match `OPENAI_API_KEY`
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')

# Import OpenAI API and set up the key
from openai import OpenAI
client = OpenAI(api_key=(OPENAI_API_KEY))
```

```python
num = input("First Number: ")
iterations = input("Second Number: ")

x = int(num)
y = int(iterations)
z = 0
for y in range(int(y)):
  z = x * x
  x = z
answer = x

response = client.responses.create(
    model="gpt-4.1-nano",
    temperature=0.5,      # try: 0.2, 0.8, 1.2, 1.6, <2.0
    top_p=0,
    input=[
    {
      "role":"developer",
      "content":"""Take only two comma seperated integers as input.
      The first number is the base. Times it by itself. It is then iterated by the
      second number of times. Display each step of the calculation to the user.
      Compare the final answer you get with the actual answer given by the user.
      If you are wrong, write a long, detailed note announcing your
      formal departure from the school of mathematics. If you are right, tell the
      user that you made a lucky guess."""
    },
    {
      "role":"assistant",
      "content":"""(2,3) will be:
                  2 * 2 = 4
                  4 * 4 = 16
                  16 * 16  = 256 """
    },
    {
      "role":"user",
      "content":str(num)+","+str(iterations)+" Answer: "+str(answer)
    }
    ]
)

print(response.output[0].content[0].text)
```
